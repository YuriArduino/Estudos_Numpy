## Escalares, Vetores e Matrizes de forma prática

### **Escalares**

* São **números individuais**, usados em cálculos simples.
* Exemplos: temperatura = 27 °C, peso = 70 kg.
* No Python/Numpy, equivalem a um **número único**.

---

### **Vetores**

* São **sequências de valores relacionados**, ou seja, listas de números que representam algo em conjunto.
* Exemplo: pesos de quatro boxeadores → `(70.0, 72.0, 71.2, 70.5)`
* Outra aplicação: posição de um ponto no espaço, como `(x, y) = (4, 3)`.
* No Numpy, **arrays 1D** são ideais para armazenar vetores.

**Resumo:** vetor = unidimensional.

---

### **Matrizes**

* Estruturas que guardam valores organizados em **linhas e colunas** (2D).
* Exemplo de tabela de dados: matriz 3x3.
* Matrizes também podem ter **mais dimensões**, como 3D ou superiores → úteis em processamento de imagens, animações e modelagem matemática.

**Resumo:**

* Dimensão indica **quantos eixos a informação varia**.

  * Vetor → 1D (uma sequência)
  * Matriz 2D → linhas × colunas
  * Matriz 3D → pilha de matrizes 2D

---

### **Importante sobre operações**

* **Soma de matrizes:** apenas se dimensões forem iguais (mesmo número de linhas e colunas).
* **Multiplicação de matrizes:** número de **colunas da primeira = número de linhas da segunda**.

---

✅ **Resumo visual mental:**

* Escalar → número único
* Vetor → lista/array 1D
* Matriz → tabela/array 2D ou multidimensional

`Teste para verificar quantidade strings`

````python
import numpy as np

url_2 = 'https://raw.githubusercontent.com/YuriArduino/Estudos_Numpy/refs/heads/Datas_tests/citrus.csv'

# 1️⃣ Lendo todos os dados como string para investigar os nomes
fruit_data_raw = np.loadtxt(url_2, delimiter=',', dtype=str, skiprows=1)

# 2️⃣ Descobrir onde termina 'orange' e onde começa 'grapefruit'
orange_end = np.where(fruit_data_raw[:, 0] != 'orange')[0][0]  # primeira linha que não é 'orange'
grapefruit_start = np.where(fruit_data_raw[:, 0] == 'grapefruit')[0][0]  # primeira linha que é 'grapefruit'


#np.where → devolve índices que atendem à condição

print("Última linha da laranja:", orange_end - 1)
print("Primeira linha da toranja:", grapefruit_start)

# 3️⃣ Agora podemos criar arrays de diâmetro e peso
# Convertendo as colunas 1 e 2 para float (diâmetro e peso)
fruit_data_numeric = fruit_data_raw[:, 1:].astype(float)

orange_diameters   = fruit_data_numeric[:orange_end, 0]  # coluna 1 -> diâmetro
orange_weights     = fruit_data_numeric[:orange_end, 1]  # coluna 2 -> peso

grapefruit_diameters = fruit_data_numeric[grapefruit_start:, 0]
grapefruit_weights   = fruit_data_numeric[grapefruit_start:, 1]

# 4️⃣ Conferir shapes
print("Shape laranjas:", orange_diameters.shape)
print("Shape toranjas:", grapefruit_diameters.shape)

#shape → devolve quantos elementos existem no array/fatia que você criou
````

Explicação:

1- Ler tudo como string (dtype=str) para poder analisar a coluna com os nomes das frutas.

2- Usar np.where para encontrar os índices de transição entre frutas.

3- Pegar apenas as colunas numéricas (diâmetro e peso) e convertendo para float com .astype(float).

4- Usar os índices encontrados para fatiar corretamente os arrays de cada fruta.


`Atenção`




O `shape` **não altera nem seleciona dados**, ele só te mostra **quantos elementos/linhas/colunas existem no array ou na fatia** que você criou.

É como um **termômetro ou conferidor**: depois que você faz um fatiamento ou filtra os dados, você pode usar `shape` para ter certeza de que o array resultante está do tamanho esperado.

Por exemplo:

```python
orange_subset = fruit_data_numeric[2000:3000, 0]
print(orange_subset.shape)  # → (1000,)
```

* Aqui o fatiamento `[2000:3000, 0]` seleciona os dados.
* O `shape` confirma que você pegou **1000 elementos**, sem precisar contar manualmente.

É como um **double-check automático** antes de trabalhar com os dados.
