## Escalares, Vetores e Matrizes de forma prtica

### **Escalares**

* So **nmeros individuais**, usados em clculos simples.
* Exemplos: temperatura = 27 C, peso = 70 kg.
* No Python/Numpy, equivalem a um **nmero nico**.

---

### **Vetores**

* So **sequncias de valores relacionados**, ou seja, listas de nmeros que representam algo em conjunto.
* Exemplo: pesos de quatro boxeadores  `(70.0, 72.0, 71.2, 70.5)`
* Outra aplicao: posio de um ponto no espao, como `(x, y) = (4, 3)`.
* No Numpy, **arrays 1D** so ideais para armazenar vetores.

**Resumo:** vetor = unidimensional.

---

### **Matrizes**

* Estruturas que guardam valores organizados em **linhas e colunas** (2D).
* Exemplo de tabela de dados: matriz 3x3.
* Matrizes tambm podem ter **mais dimenses**, como 3D ou superiores  teis em processamento de imagens, animaes e modelagem matemtica.

**Resumo:**

* Dimenso indica **quantos eixos a informao varia**.

  * Vetor  1D (uma sequncia)
  * Matriz 2D  linhas  colunas
  * Matriz 3D  pilha de matrizes 2D

---

### **Importante sobre operaes**

* **Soma de matrizes:** apenas se dimenses forem iguais (mesmo nmero de linhas e colunas).
* **Multiplicao de matrizes:** nmero de **colunas da primeira = nmero de linhas da segunda**.

---

 **Resumo visual mental:**

* Escalar  nmero nico
* Vetor  lista/array 1D
* Matriz  tabela/array 2D ou multidimensional

````python
url_2 = 'https://raw.githubusercontent.com/YuriArduino/Estudos_Numpy/refs/heads/Datas_tests/citrus.csv'
fruit_data = np.loadtxt(url_2, delimiter=',', skiprows=1, usecols=(1,2))
#np.arange(start, stop, step) [nome,dimetro,peso,vermelho,verde,azul]
#Aqui vou usar somente a um e dois [dimetro e peso]
fruit_data
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))
plt.scatter(orange_diameters, orange_weights, label='Laranja', alpha=0.3, s=7)
plt.scatter(grapefruit_diameters, grapefruit_weights, label='Toranja', alpha=0.3, s=7)


#alpha=0.5  controla a transparncia dos pontos:
#1  totalmente opaco
#0  totalmente transparente
#0.5  meio transparente, timo para ver sobreposio de pontos.
#s=10  define o tamanho dos pontos no scatter plot.
#Valores maiores  pontos maiores
#Valores menores  pontos menores

plt.xlabel('Dimetro')
plt.ylabel('Peso')
plt.title('Comparativo: Dimetro e Peso entre laranjas e toranjas')
plt.legend()
plt.grid(True)
plt.show()
````

`Teste para verificar quantidade strings`

````python
import numpy as np

url_2 = 'https://raw.githubusercontent.com/YuriArduino/Estudos_Numpy/refs/heads/Datas_tests/citrus.csv'

# 1 Lendo todos os dados como string para investigar os nomes
fruit_data_raw = np.loadtxt(url_2, delimiter=',', dtype=str, skiprows=1)

# 2 Descobrir onde termina 'orange' e onde comea 'grapefruit'
orange_end = np.where(fruit_data_raw[:, 0] != 'orange')[0][0]  # primeira linha que no  'orange'
grapefruit_start = np.where(fruit_data_raw[:, 0] == 'grapefruit')[0][0]  # primeira linha que  'grapefruit'


#np.where  devolve ndices que atendem  condio

print("ltima linha da laranja:", orange_end - 1)
print("Primeira linha da toranja:", grapefruit_start)

# 3 Agora podemos criar arrays de dimetro e peso
# Convertendo as colunas 1 e 2 para float (dimetro e peso)
fruit_data_numeric = fruit_data_raw[:, 1:].astype(float)

orange_diameters   = fruit_data_numeric[:orange_end, 0]  # coluna 1 -> dimetro
orange_weights     = fruit_data_numeric[:orange_end, 1]  # coluna 2 -> peso

grapefruit_diameters = fruit_data_numeric[grapefruit_start:, 0]
grapefruit_weights   = fruit_data_numeric[grapefruit_start:, 1]

# 4 Conferir shapes
print("Shape laranjas:", orange_diameters.shape)
print("Shape toranjas:", grapefruit_diameters.shape)

#shape  devolve quantos elementos existem no array/fatia que voc criou
````

Explicao:

1- Ler tudo como string (dtype=str) para poder analisar a coluna com os nomes das frutas.

2- Usar np.where para encontrar os ndices de transio entre frutas.

3- Pegar apenas as colunas numricas (dimetro e peso) e convertendo para float com .astype(float).

4- Usar os ndices encontrados para fatiar corretamente os arrays de cada fruta.


`Ateno`




O `shape` **no altera nem seleciona dados**, ele s te mostra **quantos elementos/linhas/colunas existem no array ou na fatia** que voc criou.

 como um **termmetro ou conferidor**: depois que voc faz um fatiamento ou filtra os dados, voc pode usar `shape` para ter certeza de que o array resultante est do tamanho esperado.

Por exemplo:

```python
orange_subset = fruit_data_numeric[2000:3000, 0]
print(orange_subset.shape)  #  (1000,)
```

* Aqui o fatiamento `[2000:3000, 0]` seleciona os dados.
* O `shape` confirma que voc pegou **1000 elementos**, sem precisar contar manualmente.

 como um **double-check automtico** antes de trabalhar com os dados.
