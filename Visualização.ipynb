```python
import numpy as np


url = 'https://raw.githubusercontent.com/YuriArduino/Estudos_Numpy/refs/heads/Datas_tests/apples_ts.csv'
dado = np.loadtxt (url, delimiter=',', usecols= np.arange(1,88,1))# np.arange(start, stop, step)
dados_transposto = dado.T
```

```python
datas = dados_transposto[:,0]
```

```python
precos = dados_transposto[:,1:6]
```

`Criando visuais para os dados`

```python
import matplotlib.pyplot as plt
plt.plot(datas,precos[:,0])
```

```python
datas
```

Datas estão sendo interpretadas como números. Ou seja, janeiro de 2013 (1.2013) está sendo lido como 1,2013.

Para solucionar este problema, adicionaremos uma célula onde a função arange() cria uma sequência com esses números, informando que correspondem aos 87 meses, portanto, de 1 a 88:

```python
datas = np.arange(1,88)
```

```python
plt.plot(datas,precos[:,0])
```

**Separando por cidades**

```python
Moscow = precos[:,0]
Kaliningrad = precos[:,1]
Petersburg = precos[:,2]
Krasnodar = precos[:,3]
Ekaterinburg = precos[:,4]
```

```python
# Teste de visualização dos preços em Moscou ao longo do tempo

plt.plot(datas, Moscow)            # plota os preços de Moscou em função das datas
plt.xlabel('Months')               # define o rótulo do eixo X como "Meses"
plt.ylabel('Apple Prices')         # define o rótulo do eixo Y como "Preços"
plt.title('Moscow', loc='center')  # adiciona título centralizado ao gráfico
```

## Escalares, Vetores e Matrizes de forma prática

### **Escalares**

* São **números individuais**, usados em cálculos simples.
* Exemplos: temperatura = 27 °C, peso = 70 kg.
* No Python/Numpy, equivalem a um **número único**.

---

### **Vetores**

* São **sequências de valores relacionados**, ou seja, listas de números que representam algo em conjunto.
* Exemplo: pesos de quatro boxeadores → `(70.0, 72.0, 71.2, 70.5)`
* Outra aplicação: posição de um ponto no espaço, como `(x, y) = (4, 3)`.
* No Numpy, **arrays 1D** são ideais para armazenar vetores.

**Resumo:** vetor = unidimensional.

---

### **Matrizes**

* Estruturas que guardam valores organizados em **linhas e colunas** (2D).
* Exemplo de tabela de dados: matriz 3x3.
* Matrizes também podem ter **mais dimensões**, como 3D ou superiores → úteis em processamento de imagens, animações e modelagem matemática.

**Resumo:**

* Dimensão indica **quantos eixos a informação varia**.

  * Vetor → 1D (uma sequência)
  * Matriz 2D → linhas × colunas
  * Matriz 3D → pilha de matrizes 2D

---

### **Importante sobre operações**

* **Soma de matrizes:** apenas se dimensões forem iguais (mesmo número de linhas e colunas).
* **Multiplicação de matrizes:** número de **colunas da primeira = número de linhas da segunda**.

---

✅ **Resumo visual mental:**

* Escalar → número único
* Vetor → lista/array 1D
* Matriz → tabela/array 2D ou multidimensional
