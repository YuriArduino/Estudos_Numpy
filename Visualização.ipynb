```python
import numpy as np


url = 'https://raw.githubusercontent.com/YuriArduino/Estudos_Numpy/refs/heads/Datas_tests/apples_ts.csv'
dado = np.loadtxt (url, delimiter=',', usecols= np.arange(1,88,1))# np.arange(start, stop, step)
dados_transposto = dado.T
```

```python
datas = dados_transposto[:,0]
```

```python
precos = dados_transposto[:,1:6]
```

`Criando visuais para os dados`

```python
import matplotlib.pyplot as plt
plt.plot(datas,precos[:,0])
```

```python
datas
```

Datas esto sendo interpretadas como nmeros. Ou seja, janeiro de 2013 (1.2013) est sendo lido como 1,2013.

Para solucionar este problema, adicionaremos uma clula onde a funo arange() cria uma sequncia com esses nmeros, informando que correspondem aos 87 meses, portanto, de 1 a 88:

```python
datas = np.arange(1,88)
```

```python
plt.plot(datas,precos[:,0])
```

## Escalares, Vetores e Matrizes de forma prtica

### **Escalares**

* So **nmeros individuais**, usados em clculos simples.
* Exemplos: temperatura = 27 C, peso = 70 kg.
* No Python/Numpy, equivalem a um **nmero nico**.

---

### **Vetores**

* So **sequncias de valores relacionados**, ou seja, listas de nmeros que representam algo em conjunto.
* Exemplo: pesos de quatro boxeadores  `(70.0, 72.0, 71.2, 70.5)`
* Outra aplicao: posio de um ponto no espao, como `(x, y) = (4, 3)`.
* No Numpy, **arrays 1D** so ideais para armazenar vetores.

**Resumo:** vetor = unidimensional.

---

### **Matrizes**

* Estruturas que guardam valores organizados em **linhas e colunas** (2D).
* Exemplo de tabela de dados: matriz 3x3.
* Matrizes tambm podem ter **mais dimenses**, como 3D ou superiores  teis em processamento de imagens, animaes e modelagem matemtica.

**Resumo:**

* Dimenso indica **quantos eixos a informao varia**.

  * Vetor  1D (uma sequncia)
  * Matriz 2D  linhas  colunas
  * Matriz 3D  pilha de matrizes 2D

---

### **Importante sobre operaes**

* **Soma de matrizes:** apenas se dimenses forem iguais (mesmo nmero de linhas e colunas).
* **Multiplicao de matrizes:** nmero de **colunas da primeira = nmero de linhas da segunda**.

---

 **Resumo visual mental:**

* Escalar  nmero nico
* Vetor  lista/array 1D
* Matriz  tabela/array 2D ou multidimensional
